Bu yüzden görselleştirme için yeni bir dataframe oluştururken bu sütun kaldırılır.

```python
df1= df.drop(["Id"],axis=1)
```

```python
df1
```

```python
df1.plot()
plt.show()  #çizilen plotun gösterilmesini sağlar.
#burada hangisinin versicolor, virginica olduğu belli değildir, tüm data çizdirilmiştir. 
#ancak keskin geçişlerin olduğu yerler tür değişimleri olabilir ve her bir tür için 50 sample olduğu bilgisini biliyoruz. 
```

```python
setosa = df[ df.Species == "Iris-setosa"] 
#setosa türünün ayrıldığı bir dataframe oluşturuldu.

versicolor = df[ df.Species == "Iris-versicolor"] 
#versicolor türünün ayrıldığı bir dataframe oluşturuldu.

virginica = df[ df.Species == "Iris-virginica"] 
#virginica türünün ayrıldığı bir dataframe oluşturuldu.
```

hangi türün daha iyi olduğunu anlamak için ayrı ayrı plot edip tek bir plot üzerinde gösterilebilir.

```python
plt.plot(setosa.Id, setosa.PetalLengthCm, color="green", label = "setosa")
plt.plot(versicolor.Id, versicolor.PetalLengthCm, color="blue", label = "versicolor")
plt.plot(virginica.Id, virginica.PetalLengthCm, color="red", label = "virginica")

plt.legend() #legend methoduyla label eklendi, methodun içinin default bırakılmasının sebebi grafikte en uygun yere koyulması içindir.
plt.xlabel("Id") #x eksenine isim verildi
plt.ylabel("PetalLengthCm") #y eksenine isim verildi.
plt.show()
```

```python
df1.plot(grid=True) #grid ile grafik çizilirken dikdörtgen olarak ayırır, böylece incelemek kolaylaşır.
plt.show()
```

```python
df1.plot(grid=True, linestyle = ":") #linestyle grafiğin çizim şeklini değiştirilebilir.
plt.show()
```

```python
df1.plot(grid=True, alpha=0.5) #alpha ile çizgilerin saydamlığı artırılır, 0'a ne kadar yaklaşırsa o kadar artar.
plt.show()
```
