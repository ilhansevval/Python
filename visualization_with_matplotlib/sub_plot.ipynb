```python
setosa = df[ df.Species == "Iris-setosa"] 
#setosa türünün ayrıldığı bir dataframe oluşturuldu.

versicolor = df[ df.Species == "Iris-versicolor"] 
#versicolor türünün ayrıldığı bir dataframe oluşturuldu.

virginica = df[ df.Species == "Iris-virginica"] 
#virginica türünün ayrıldığı bir dataframe oluşturuldu.
```

```python
plt.subplot(3,1,1) #3 subplottan 1.columnun birinci satırı
plt.plot(setosa.Id, setosa.PetalLengthCm, color="green", label = "setosa")
plt.ylabel("setosa")

plt.subplot(3,1,2) #3 subplottan 1.columnun ikinci satırı
plt.plot(versicolor.Id, versicolor.PetalLengthCm, color="blue", label = "versicolor")
plt.ylabel("versicolor")

plt.subplot(3,1,3) #3 subplottan 1.columnun üçüncü satırı
plt.plot(virginica.Id, virginica.PetalLengthCm, color="red", label = "virginica")
plt.ylabel("virginica")

plt.show()
```
